```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Dataset cargado
ruta = "/content/Espisodios.csv"
df = pd.read_csv(ruta)
df.head(5)
```

```python
#Dataset N2
ruta2 = "/content/series.csv"
df2 = pd.read_csv(ruta2)
df2.head(5)
```

```python
df.describe()
```

```python
#Gráfico Nª1
promedio_rating_por_serie = df.groupby('serie_id')['rating_imdb'].mean().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='serie_id', y='rating_imdb', data=promedio_rating_por_serie, palette='viridis', hue='serie_id', legend=False)
plt.title('Promedio de Rating IMDb por Serie')
plt.xlabel('ID de la Serie')
plt.ylabel('Promedio de Rating IMDb')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)


for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
#Gráfico Nª2
df['fecha_estreno'] = pd.to_datetime(df['fecha_estreno'])
df['año_estreno'] = df['fecha_estreno'].dt.year

plt.figure(figsize=(12, 7))
sns.histplot(data=df, x='rating_imdb', hue='año_estreno', multiple='dodge', bins=10, palette='viridis')
plt.title('Distribución del Rating IMDb por Año de Estreno')
plt.xlabel('Rating IMDb')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Gráfico Nª3
promedio_duracion_por_serie = df.groupby('serie_id')['duracion'].mean().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.pointplot(x='serie_id', y='duracion', data=promedio_duracion_por_serie, palette='viridis', hue='serie_id', legend=False, marker='*', linestyles='')
plt.title('Promedio de Duración por Serie (Point Plot)')
plt.xlabel('ID de la Serie')
plt.ylabel('Promedio de Duración')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.9)

for i, row in promedio_duracion_por_serie.iterrows():
    ax.annotate(f'{row['duracion']:.2f}',
                (i, row['duracion']),
                textcoords="offset points",
                xytext=(0,10), # distance from text to points (x,y)
                ha='center')

plt.tight_layout()
plt.show()
```
